```python
# To write the suitable code for a C program that can find the sum and average of n given numbers

n=int(input("Type in how many numbers you want to sum and average >>> "))
array = []
for i in range(0,n): # for loop over a sequence generated by the function range
    x = int(input("Enter the number >>> "))
    array.append(x)
print('The sum of the numbers is ',sum(array), 'and the average is', (sum(array)/n)) #sum() function adds all the elements 
```

```python
# To find the maximum among the numbers
print('The maximum among them is',max(array))
```

```python
# To sort the numbers
print('The numbers sorted in ascending order are as follows:\n',sorted(array, reverse=False))
print('The numbers sorted in descending order are as follows:\n',sorted(array, reverse=True))
```
